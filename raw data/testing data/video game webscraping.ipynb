```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup as soup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
browser = Browser('chrome')
```

```python
# Visit the website
url = "https://www.vgchartz.com/yearly/2018/Global/"
browser.visit(url)
```

```python
# Create a Beautiful Soup Object
html= browser.html
global_2018_soup = soup(html,"html.parser")
global_2018_soup
```

```python
# Extract Video Game Names
video_game_names = global_2018_soup.find_all("td")
video_game_names
```

```python
first_thing = global_2018_soup.find("td")
first_thing_test = first_thing.renderContents()
first_thing_test
```
