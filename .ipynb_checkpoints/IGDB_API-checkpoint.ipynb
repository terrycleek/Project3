```python
# Define platform IDs for different platforms (replace with actual IDs)
pc_platform_id = 6
ps5_platform_id = 167
xbox_platform_id = 169
switch_platform_id = 130

# Define the API endpoint and query parameters
url = "https://api.igdb.com/v4/games"
params = {
    "fields": "name,platforms",
    "limit": 10,  # Limit the results to 10 games
}

# Make a GET request to fetch game data
response = requests.get(url, headers=headers, params=params)

if response.status_code == 200:
    game_data = response.json()

    # Separate data for different platforms
    pc_games = [game for game in game_data if pc_platform_id in game.get("platforms", [])]
    ps5_games = [game for game in game_data if ps5_platform_id in game.get("platforms", [])]
    xbox_games = [game for game in game_data if xbox_platform_id in game.get("platforms", [])]
    switch_games = [game for game in game_data if switch_platform_id in game.get("platforms", [])]

    # Display or use the separated data
    print("PC Games:", pc_games)
    print("PlayStation 5 Games:", ps5_games)
    print("Xbox Series X/S Games:", xbox_games)
    print("Nintendo Switch Games:", switch_games)
else:
    print(f"Error: {response.status_code}")
```

```python
for console in console_data:
    platform_id = console['id']
    
    # Get games for a specific console
    response = requests.get(
        "https://api.igdb.com/v4/games",
        headers=headers,
        params={"fields": "name,genres", "where": f"platforms={platform_id}"}
    )
    
    if response.status_code == 200:
        games_data = response.json()
        # Now, you have games for a specific console in games_data
        # You can further process this data, including sorting by genres
    else:
        print(f"Error: {response.status_code}")
```

```python
# Example code to group games by genre
games_by_genre = {}

for game in games_data:
    for genre_id in game.get('genres', []):
        genre_name = genre_id  # Replace with actual logic to get genre names
        if genre_name not in games_by_genre:
            games_by_genre[genre_name] = []
        games_by_genre[genre_name].append(game['name'])
```
